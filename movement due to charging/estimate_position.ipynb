```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w0 = 2.5e-6 # waist in meters
N_pi = np.linspace(100, 500, 1000)
x = w0 / np.sqrt(N_pi)
plt.plot(N_pi, x * 1e6)
plt.xlabel("Tau / pi time")
plt.ylabel("Estimated displacement (um)")
plt.title("Estimated shot-to-shot displacement to explain decoherence")
plt.grid()
```
